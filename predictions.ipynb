```python
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
model_pred = tf.keras.models.load_model('Final Model.h5')
```

```python
model_pred.compile(loss='categorical_crossentropy',
             optimizer='adam',
                metrics=['accuracy'])
```

```python
class_names = ['Building','Forest','Galcier','Mountain','Sea','Street']

def image_prediction(filename):
    img = cv2.imread(filename)
    imgPlot = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(14,8))
    plt.imshow(imgPlot)
    plt.axis('off')
    img = cv2.resize(img,(150,150))
    img = np.reshape(img,[1,150,150,3])
    img = img.astype('float64')
    img = img / 255.0
    pred = model_pred.predict(img)
    name = class_names[np.argmax(pred)]
    plt.title('This is predicted as ' + name,size=25)
```

```python
image_prediction('image1.jpg')
```

```python
image_prediction('image2.jpg')
```
